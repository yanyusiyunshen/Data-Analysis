# Pandas的两种数据类型

## Series

Pandas有两个最主要也是最重要的数据结构:**Series**和**DataFrame**Series是一种类似于一维数组的**对象**，由一组数据(各种NumPy数据类型)以及一组与之对应的索引(数据标签)组成。

- 类似一维数组的对象

- 由数据和索引组成

  - 索引(index)在左，数据(values)在右

  - 索引是自动创建的

### 1.通过list创建Series

```python
import pandas as pd
# ser_obj = pd.Series(range(10))
#通过list构建
# print(ser_obj)
ser_obj =pd.Series (range(10, 20))

print(ser_obj.head(3))
print('-'*20)
print(ser_obj)
print('-'*20)
print(type(ser_obj))
print('\n')
# print(Series_obj.head(3))
# print('-'*20)
# print(Series_obj)
# print('-'*20)
# print(type(Series_obj))
```

### 3.通过索引获得数据

## DataFrame

<br>DataFrame是一个表格型的数据结构，它含有一组有序的列，每列可以是不同型的值。DataFrame既有行索引也有列索引，它可以被看做是由Series组成的字典(共用同一个索引)，数据是以二维结构存放的。
- 类似多维数组/表格数据(如，excel,R中的data.frame)
- 每列数据可以是不同的类型
- 索引包括列索引和行索引

每列数据可以是不同的类型- 索引包括列索引和行索引DataFrame对象既有行索引，又有列索引

行索引，表明不同行，横向索引，叫index，0轴，axis=0

列索引，表名不同列，纵向索引，叫columns，1轴，axis=1

# Pandas的对齐运算

是数据清洗的重要过程，可以按索引对齐进行运算，如果没对齐的位置则补NaN，最后也可以填充NaN

```python
a=np.array([1,2,3,4,5])
b=np.array([1])
print(a+b)
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3])
print(a+b)
```

```python
import pandas as pd
s1 = pd.Series(range(10, 20), index = range(10))
s2 = pd.Series(range(20, 25), index = range(5))
# Series 对齐运算
print('sl+s2: ')
s3=s1+s2
print(s3)#缺失数据默认是NaN np.nan
```

```python
print(np.isnan(s3[6]))
print('-'*20)

print(s3)
print('-'*20)

print(s1)
print('-'*20)

print(s2)
print('-'*20)

print(s2.add(s1, fill_value = 0)) #未对齐的数据将和填充值做运算
print('-'*20)

print(s2.sub(s1, fill_value = 0))
```

```python
#df的对齐运算
import numpy as np
df1 = pd.DataFrame(np.ones((2,2)),columns = ['a', 'b'])
df2 = pd.DataFrame(np.ones((3,3)), columns = ['a', 'b', 'c'])
print(df1)
print(df2)
print('-'*50)
print(df2.dtypes)
print(df1-df2)
print(df2.sub(df1,fill_value=2))#未对齐的数据将和填充值做运算
```

## 1.Pandas的索引排序

```python
# Series
import pandas as pd
import numpy as np
print(np.random.randint(5, size=5))
s4 = pd.Series(range(10, 15), index = np.random.randint(5,size=5)) #索引随机生成
print(s4)
print('-'*50)
#索引排序,sort_index返回一个新的排好索引的series
print(s4.sort_index())
print('-'*50)
print(s4)
#s4.loc[0:3] loc索引值不唯一时直接报错
print(s4.iloc[0:3])
print('-'*50)
s4[0:3] #默认用的位置索引
```

```python
# DataFrame
df4 = pd.DataFrame (np.random.randn(5, 5),
                    index=np.random.randint(5, size=5),
                    columns=np.random.randint(5, size=5))
print(df4)
#轴零是行索引排序
df4_isort = df4.sort_index(axis=0, ascending=False)
print(df4_isort)
print('-'*50)

df4_isort = df4.sort_index(axis=0, ascending=True)
print(df4_isort)
print('-'*50)

df4_isort = df4.sort_index(axis=1, ascending=True)
print(df4_isort)
```

# 层级索引

```python
import pandas as pd
import numpy as np
#MultiIndex是层级索引，索引类型的一种
index1 = pd.MultiIndex.from_arrays([['a','a','a','b','b','b','c','c','c','d','d','d'],
                                  [0, 1, 2, 0, 1, 2, 0, 1, 2, 0, 1, 2]], names=['cloth', 'size'])
ser_obj = pd.Series(np.random.randn(12),index = index1)
print(ser_obj)
print('-'*20)
print(type(ser_obj)) #Series
print('-'*20)
print(type(ser_obj.index)) #索引类型，MultiIndex
print('-'*20)
print(ser_obj.index)
print('-'*20)
print(ser_obj.index.levels) #层级索引的索引值
print('-'*20)
ser_obj.index.codes #没那么重要，代表索引的位置
```

```python
#层级索引如何取数据
print(ser_obj.loc['c'])#取出c的所有数据，取出的是series
print('-'*20)
print(ser_obj['c'])#取出c的所有数据，取出的是series
print('-'*20)
print(ser_obj['a', 2])
print('-'*20)
print(ser_obj[:,2])#取出所有行的内层索引为2的数据
```

交换层级索引

```python
print(ser_obj.swaplevel())
print('-'*20)
print(ser_obj)
print('-'*20)
ser_obj=ser_obj.swaplevel()
print(ser_obj)
```

```python
#层级索引按那个索引级别排序，level=0表示按最外层索引排序
print(ser_obj.sort_index(level=1))
print('-'*20)
print(ser_obj.sort_index(level=0))
```

```python
# 把O索引(最外层索引)变为列索引
print(ser_obj.sort_index(level=0))
print('-'*20)
df_obj=ser_obj.unstack(0)   # unstack可以放索引名，或者索引位置
print(df_obj)
```

```python
print(df_obj)
#对df进行stack，就会把行，列索引进行堆叠，变为series
#把列索引放入内层，只能放到内层
print(df_obj.stack()) #stack变为series和unstack保持一致的
# df_obj=df_obj.transpose
```

# 分组和聚合

## 分组

```python
import pandas as pd
import numpy as np
dict_obj = {'key1':['a','b','a','b','a','b','a','a'],
            'key2':['one','one','two','three','two','two','one','three'],
            'data1':np.random.randn(8),
            'data2':np.random.randn(8)}
df_obj = pd.DataFrame(dict_obj)
print(df_obj)#两列是字符串，两列是数值
```

```python
#dataframe根据key1进行分组，分组后的对象类型#不进行聚合操作拿到的只是groupby的对象
print(df_obj.groupby('key1'))
print(type(df_obj.groupby('key1')))
```

```python
group1=df_obj.groupby('key1')
for name,group in group1:#遍历分组
    print(name)
    print(group)
```

```python
group2=df_obj.groupby('key2')
for name,group in group2:#遍历分组
    print(name)
    print(group)
```

```python
print(df_obj['data1']) #dataframe的 data1 列根据 key1 进行分组
print('-'*50)
print(type(df_obj['data1'].groupby(df_obj['key1'])))
```

```python
#分组运算
grouped1 = df_obj.groupby('key1')
print(grouped1.mean(numeric_only=True))#不是数值类型的不会算均值

print('-'*50)

grouped2 = df_obj['data1'].groupby(df_obj['key1'])
print(grouped2.mean())
```

## 聚合

```python
import pandas as pd
import numpy as np
```

```python
dict_obj = {'key1': ['a', 'b','a','b','a','b','a','a'],
            # 'key2':['one','one','two', 'three','two','two','one','three'],
            'data1': np.random.randint(1,10, 8),
            'data2': np.random.randint(1,10, 8)}
df_obj=pd.DataFrame(dict_obj)
print(df_obj.loc[:,['key1','data1','data2']].groupby("key1").sum())
print(df_obj.loc[:,['key1','data1','data2']].groupby("key1").max())
```

```python
print(df_obj.loc[:,['key1','data1','data2']].groupby("key1").min())
print('-'*50)#平均值
print(df_obj.loc[:, ['key1','data1','data2']].groupby("key1").mean())
print('-'*50)
df_obj.groupby("key1").describe()
```

### 自定义聚合

```python


df_obj5 = pd.DataFrame (dict_obj)
print(df_obj5)
def peak_range(df):
    """
    返回数值范围，df参数是groupby后的 那一组的dataframe
    """
    #print type(df)#参数为索引所对应的记录
    return df.max()-df.min()
#求极值
print('-'*50)
print(df_obj5.loc[:, ['key1','data1','data2']].groupby('key1').agg (peak_range))
print('-'*50)
print(df_obj5.loc[:, ['key1','data1', 'data2']].groupby('key1').agg(lambda df : df. max() - df.min()))
```

```python
#同时应用多个聚合函数
# 只选择数值列
numeric_cols = df_obj.select_dtypes(include=['number']).columns
print(df_obj.groupby('key1')[numeric_cols].agg(['mean', 'std', 'count',peak_range]))
# print(df_obj.groupby('key1').agg(['mean', 'std', 'count', peak_range]))
```

```python
#每列作用不同的聚合函数
dict_mapping={'data1':'mean',
              'data2':'count'}
print(df_obj.groupby('key1').agg(dict_mapping))
```
