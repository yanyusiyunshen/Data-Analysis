# pandas的两种数据类型

## Series

Pandas有两个最主要也是最重要的数据结构:**Series**和**DataFrame**Series是一种类似于一维数组的**对象**，由一组数据(各种NumPy数据类型)以及一组与之对应的索引(数据标签)组成。

- 类似一维数组的对象

- 由数据和索引组成

  - 索引(index)在左，数据(values)在右

  - 索引是自动创建的

### 1.通过list创建Series

```python
import pandas as pd
# ser_obj = pd.Series(range(10))
#通过list构建
# print(ser_obj)
ser_obj =pd.Series (range(10, 20))

print(ser_obj.head(3))
print('-'*20)
print(ser_obj)
print('-'*20)
print(type(ser_obj))
print('\n')
# print(Series_obj.head(3))
# print('-'*20)
# print(Series_obj)
# print('-'*20)
# print(type(Series_obj))
```

### 2.获取数据和索引

```python
# ser_obj.index 和ser_obj.values

#获取数据
print(ser_obj.values)
#获取索引
print(ser_obj.index)
```

### 3.通过索引获得数据

```python
# ser_obj[idx]

#通过索引获取数据
print(ser_obj[0])
print(ser_obj[8])
```

### 4.索引与数据的对应关系不被运算结果影响

```python
#索引与数据的对应关系不被运算结果影响
print(ser_obj * 2)
print(ser_obj > 15)
```

### 5.通过dict构建Series

```python
#通过dict构建Series
year_data = {2001: 17.8, 2002: 20.1, 2003: 16.5}
ser_obj2 = pd.Series(year_data)
print(ser_obj2.head())
print(ser_obj2.index)
print (ser_obj2[2001])
```

### 6.name属性

```python
# 对象名:ser_obj.name
# 对象索引名:ser_obj.index.name

# name属性
ser_obj2.name = 'temp'
ser_obj2.index.name = 'year'
print(ser_obj2.head())
```

## DataFrame

<br>DataFrame是一个表格型的数据结构，它含有一组有序的列，每列可以是不同型的值。DataFrame既有行索引也有列索引，它可以被看做是由Series组成的字典(共用同一个索引)，数据是以二维结构存放的。
- 类似多维数组/表格数据(如，excel,R中的data.frame)
- 每列数据可以是不同的类型
- 索引包括列索引和行索引

每列数据可以是不同的类型- 索引包括列索引和行索引DataFrame对象既有行索引，又有列索引

行索引，表明不同行，横向索引，叫index，0轴，axis=0

列索引，表名不同列，纵向索引，叫columns，1轴，axis=1

###  Series索引

#### 1.index指定行索引名

```python
import pandas as pd
ser_obj = pd.Series (range(5), index = ['a', 'b','c','d','e'])
print(ser_obj.head())

ser_obj1 = pd. Series (range (5), index = list ("abcde"))
print (ser_obj1)
```

#### 2.行索引

```python
# ser_obj['label‘], ser_obj[pos]
#行索引
print(ser_obj.loc['b']) # 索引名
print(ser_obj.iloc[2])  # 索引位置
```

#### 3.切片索引

```python
#切片索引
#索引位置取数据，左闭右开
print(ser_obj.iloc[1:3])
#记住索引名左闭右闭
print(ser_obj.loc['b':'d'])
```

#### 4.不连续索引

```python
print(ser_obj.iloc[[0, 2, 4]])
print(ser_obj.loc[['a', 'e']])
```

#### 5.布尔索引

```python
ser_bool = ser_obj > 2
print(ser_obj)
print(ser_bool)
print('-'*20)
print(ser_obj[ser_bool])
print(ser_obj[ser_obj>2])#取出大于2的元素
```

### DataFrame索引

#### 1.columns指定列索引名

```python
import numpy as np
df_obj = pd.DataFrame(np.random.randn(5,4),
columns= ['a','b','c','d'])
print(df_obj.head())
```

#### 2.列索引

```python
#列索引
print (df_obj['a']) # 返回Series类型
print (df_obj[['a']]) # 返回DataFrame类型
print (type (df_obj[['a']])) # 返回DataFrame 类型
```
