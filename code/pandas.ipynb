# pandas的两种数据类型

Pandas有两个最主要也是最重要的数据结构:**Series**和**DataFrame**Series是一种类似于一维数组的**对象**，由一组数据(各种NumPy数据类型)以及一组与之对应的索引(数据标签)组成。

- 类似一维数组的对象

- 由数据和索引组成

  - 索引(index)在左，数据(values)在右

  - 索引是自动创建的

## 1.通过list创建Series

```python
import pandas as pd
# ser_obj = pd.Series(range(10))
#通过list构建
# print(ser_obj)
ser_obj =pd.Series (range(10, 20))

print(ser_obj.head(3))
print('-'*20)
print(ser_obj)
print('-'*20)
print(type(ser_obj))
print('\n')
# print(Series_obj.head(3))
# print('-'*20)
# print(Series_obj)
# print('-'*20)
# print(type(Series_obj))
```

## 2.获取数据和索引

```python
# ser_obj.index 和ser_obj.values

#获取数据
print(ser_obj.values)
#获取索引
print(ser_obj.index)
```

## 3.通过索引获得数据

```python
# ser_obj[idx]

#通过索引获取数据
print(ser_obj[0])
print(ser_obj[8])
```

## 4.索引与数据的对应关系不被运算结果影响

```python
#索引与数据的对应关系不被运算结果影响
print(ser_obj * 2)
print(ser_obj > 15)
```

## 5.通过dict构建Series

```python
#通过dict构建Series
year_data = {2001: 17.8, 2002: 20.1, 2003: 16.5}
ser_obj2 = pd.Series(year_data)
print(ser_obj2.head())
print(ser_obj2.index)
print (ser_obj2[2001])
```

## 6.name属性

```python
# 对象名:ser_obj.name
# 对象索引名:ser_obj.index.name

# name属性
ser_obj2.name = 'temp'
ser_obj2.index.name = 'year'
print(ser_obj2.head())
```
