# Pandas的两种数据类型

## Series

Pandas有两个最主要也是最重要的数据结构:**Series**和**DataFrame**Series是一种类似于一维数组的**对象**，由一组数据(各种NumPy数据类型)以及一组与之对应的索引(数据标签)组成。

- 类似一维数组的对象

- 由数据和索引组成

  - 索引(index)在左，数据(values)在右

  - 索引是自动创建的

### 1.通过list创建Series

```python
import pandas as pd
# ser_obj = pd.Series(range(10))
#通过list构建
# print(ser_obj)
ser_obj =pd.Series (range(10, 20))

print(ser_obj.head(3))
print('-'*20)
print(ser_obj)
print('-'*20)
print(type(ser_obj))
print('\n')
# print(Series_obj.head(3))
# print('-'*20)
# print(Series_obj)
# print('-'*20)
# print(type(Series_obj))
```

### 2.获取数据和索引

```python
# ser_obj.index 和ser_obj.values

#获取数据
print(ser_obj.values)
#获取索引
print(ser_obj.index)
```

### 3.通过索引获得数据

```python
# ser_obj[idx]

#通过索引获取数据
print(ser_obj[0])
print(ser_obj[8])
```

### 4.索引与数据的对应关系不被运算结果影响

```python
#索引与数据的对应关系不被运算结果影响
print(ser_obj * 2)
print(ser_obj > 15)
```

### 5.通过dict构建Series

```python
#通过dict构建Series
year_data = {2001: 17.8, 2002: 20.1, 2003: 16.5}
ser_obj2 = pd.Series(year_data)
print(ser_obj2.head())
print(ser_obj2.index)
print (ser_obj2[2001])
```

### 6.name属性

```python
# 对象名:ser_obj.name
# 对象索引名:ser_obj.index.name

# name属性
ser_obj2.name = 'temp'
ser_obj2.index.name = 'year'
print(ser_obj2.head())
```

## DataFrame

<br>DataFrame是一个表格型的数据结构，它含有一组有序的列，每列可以是不同型的值。DataFrame既有行索引也有列索引，它可以被看做是由Series组成的字典(共用同一个索引)，数据是以二维结构存放的。
- 类似多维数组/表格数据(如，excel,R中的data.frame)
- 每列数据可以是不同的类型
- 索引包括列索引和行索引

每列数据可以是不同的类型- 索引包括列索引和行索引DataFrame对象既有行索引，又有列索引

行索引，表明不同行，横向索引，叫index，0轴，axis=0

列索引，表名不同列，纵向索引，叫columns，1轴，axis=1

# Pandas的索引操作

## 1.Series和DataFrame中的索引都是Index对象

###  Series索引

#### 1.index指定行索引名

```python
import pandas as pd
ser_obj = pd.Series (range(5), index = ['a', 'b','c','d','e'])
print(ser_obj.head())

ser_obj1 = pd. Series (range (5), index = list ("abcde"))
print (ser_obj1)
```

#### 2.行索引

```python
# ser_obj['label‘], ser_obj[pos]
#行索引
print(ser_obj.loc['b']) # 索引名
print(ser_obj.iloc[2])  # 索引位置
```

#### 3.切片索引

```python
#切片索引
#索引位置取数据，左闭右开
print(ser_obj.iloc[1:3])
#记住索引名左闭右闭
print(ser_obj.loc['b':'d'])
```

#### 4.不连续索引

```python
print(ser_obj.iloc[[0, 2, 4]])
print(ser_obj.loc[['a', 'e']])
```

#### 5.布尔索引

```python
ser_bool = ser_obj > 2
print(ser_obj)
print(ser_bool)
print('-'*20)
print(ser_obj[ser_bool])
print(ser_obj[ser_obj>2])#取出大于2的元素
```

### DataFrame索引

#### 1.columns指定列索引名

```python
import numpy as np
df_obj = pd.DataFrame(np.random.randn(5,4),
columns= ['a','b','c','d'])
print(df_obj.head())
```

#### 2.列索引

```python
#列索引
print (df_obj['a']) # 返回Series类型
print (df_obj[['a']]) # 返回DataFrame类型
print (type (df_obj[['a']])) # 返回DataFrame 类型
```

```python
#列索引
print(df_obj.loc[:,'a']) # 返回Series类型print('-'*50)
print(df_obj.loc[:,['a']]) #返回DataFrame类型print('-'*50)
print(type(df_obj.loc[:,['a']])) #返回DataFrame类型
```

```python
print(df_obj.loc[1,['a']])
# 取一行
print(df_obj.loc[1,:])
```

#### 3.不连续索引

```python
#标签索loc，建议使用Loc，效率更高#Series
print(ser_obj)
print(ser_obj.loc['b':'d'])#前闭后闭
# print('-'*50)
```

```python
# DataFrame
df_obj = pd.DataFrame (np.random.randn(5,4),columns = list('abcd'),index=list('abcde'))
print(df_obj)
print('-'*20)

print(df_obj.loc['a']) #拿的是行
print('-'*20)

# 第一个参数索引行，第二个参数是列，loc或者iloc效率高于直接用取下标的方式，前闭后闭
print(df_obj.loc['a':'c','b':'d'])
print('-'*20)

#连续索引
print(df_obj.loc[['a','c'],['b','d']])
print('-'*20)

#不连续索引
print(df_obj.loc[['c'],['b']])
print('-'*20)

#取一个值，返回的是DataFrame类型
print(df_obj.loc['c','b']) #取一个值
```

#### 4.loc标签索引

```python
# DataFrame
df_obj = pd.DataFrame (np.random.randn(5,4),columns = list('abcd'),index=list('abcde'))
print(df_obj)
print('-'*20)

print(df_obj.loc['a']) #拿的是行
print('-'*20)

# 第一个参数索引行，第二个参数是列，loc或者iloc效率高于直接用取下标的方式，前闭后闭
print(df_obj.loc['a':'c','b':'d'])
print('-'*20)

#连续索引
print(df_obj.loc[['a','c'],['b','d']])
print('-'*20)

#不连续索引
print(df_obj.loc[['c'],['b']])
print('-'*20)

#取一个值，返回的是DataFrame类型
print(df_obj.loc['c','b']) #取一个值
```

#### 5.iloc位置索引

```python
# Series
print(ser_obj[1:3])
print(ser_obj.iloc[1:3]) # 前闭后开[)，效率高print('-'*50)
print('-'*20)

#DataFrame，iloc是前闭后开[)
print(df_obj)
print('-'*20)

print(df_obj.iloc[0:2,0:2]) # 注意和df_obj.loc[0:2,'a']的区别
print('-'*20)

print(df_obj.iloc[[0,2],[0,2]])
print('-'*20)
# 不连续索引
print(df_obj.iloc[0,0]) # 取一个值
```

# Pandas的对齐运算

是数据清洗的重要过程，可以按索引对齐进行运算，如果没对齐的位置则补NaN，最后也可以填充NaN

```python
a=np.array([1,2,3,4,5])
b=np.array([1])
print(a+b)
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3])
print(a+b)
```

```python
import pandas as pd
s1 = pd.Series(range(10, 20), index = range(10))
s2 = pd.Series(range(20, 25), index = range(5))
# Series 对齐运算
print('sl+s2: ')
s3=s1+s2
print(s3)#缺失数据默认是NaN np.nan
```

```python
print(np.isnan(s3[6]))
print('-'*20)

print(s3)
print('-'*20)

print(s1)
print('-'*20)

print(s2)
print('-'*20)

print(s2.add(s1, fill_value = 0)) #未对齐的数据将和填充值做运算
print('-'*20)

print(s2.sub(s1, fill_value = 0))
```

```python
#df的对齐运算
import numpy as np
df1 = pd.DataFrame(np.ones((2,2)),columns = ['a', 'b'])
df2 = pd.DataFrame(np.ones((3,3)), columns = ['a', 'b', 'c'])
print(df1)
print(df2)
print('-'*50)
print(df2.dtypes)
print(df1-df2)
print(df2.sub(df1,fill_value=2))#未对齐的数据将和填充值做运算
```

# Pandas的函数应用

## 1.Pandas的索引排序

```python
# Series
import pandas as pd
import numpy as np
print(np.random.randint(5, size=5))
s4 = pd.Series(range(10, 15), index = np.random.randint(5,size=5)) #索引随机生成
print(s4)
print('-'*50)
#索引排序,sort_index返回一个新的排好索引的series
print(s4.sort_index())
print('-'*50)
print(s4)
#s4.loc[0:3] loc索引值不唯一时直接报错
print(s4.iloc[0:3])
print('-'*50)
s4[0:3] #默认用的位置索引
```

```python
# DataFrame
df4 = pd.DataFrame (np.random.randn(5, 5),
                    index=np.random.randint(5, size=5),
                    columns=np.random.randint(5, size=5))
print(df4)
#轴零是行索引排序
df4_isort = df4.sort_index(axis=0, ascending=False)
print(df4_isort)
print('-'*50)

df4_isort = df4.sort_index(axis=0, ascending=True)
print(df4_isort)
print('-'*50)

df4_isort = df4.sort_index(axis=1, ascending=True)
print(df4_isort)
```
